# ğŸŒ¸ â€” Flowers Classification with MobileNetV2

## 1ï¸âƒ£ Introduction âœ¨
In this project, we build an **image classifier using Transfer Learning** with **MobileNetV2**.  
The model predicts **5 flower categories**:

- ğŸŒ¼ daisy  
- ğŸŒ» sunflower  
- ğŸŒ¹ rose  
- ğŸŒ· tulip  
- ğŸŒ¾ dandelion  

This project demonstrates:

âœ… fine-tuning

âœ… Grad-CAM interpretability

## 2ï¸âƒ£ Imports & Dataset Paths ğŸ“¦
We load the required libraries and define dataset paths.


```python
import os
import cv2,random
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import random

from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# dataset path
DATASET_PATH = "data/flowers"
print("Classes:", os.listdir(DATASET_PATH))
```

## 3ï¸âƒ£ Data Loading & Augmentation ğŸŒ€
We apply augmentation only on the training set to improve generalization.


```python
IMG_SIZE = (160, 160)
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=25,
    zoom_range=0.15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    fill_mode="nearest"
)

train_gen = train_datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training",
    shuffle=True
)

val_gen = train_datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=False
)

num_classes = train_gen.num_classes
class_names = list(train_gen.class_indices.keys())
print("Class names:", class_names)
```

## 4ï¸âƒ£ Transfer Learning Setup ğŸ§Š
We load **MobileNetV2 pretrained on ImageNet** and freeze all convolutional layers.  
Then we add a custom classification head.


```python
base_model = MobileNetV2(
    input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3),
    include_top=False,
    weights="imagenet"
)

base_model.trainable = False  # freeze backbone

inputs = Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
x = base_model(inputs, training=False)
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)
outputs = Dense(num_classes, activation="softmax")(x)

model = Model(inputs, outputs)

model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

## 5ï¸âƒ£ Training Stage 1 (Frozen Backbone) ğŸ‹ï¸â€â™€ï¸
We train only the new classification head.


```python
callbacks = [
    EarlyStopping(patience=4, restore_best_weights=True, monitor="val_loss"),
    ModelCheckpoint("models/best_model_stage1.keras", save_best_only=True, monitor="val_loss")
]

history1 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,
    callbacks=callbacks
)
```

## 6ï¸âƒ£ Fine-Tuning Stage 2 ğŸ”¥
We unfreeze the top part of MobileNetV2 and continue training with a smaller LR.


```python
base_model.trainable = True

# freeze lower layers, unfreeze top ~30 layers
for layer in base_model.layers[:-30]:
    layer.trainable = False

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

callbacks_ft = [
    EarlyStopping(patience=4, restore_best_weights=True, monitor="val_loss"),
    ModelCheckpoint("models/best_model_finetuned.keras", save_best_only=True, monitor="val_loss")
]

history2 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,
    callbacks=callbacks_ft
)
```

## 7ï¸âƒ£ Training Curves ğŸ“ˆ
We visualize accuracy and loss across both training stages.


```python
def plot_history(h1, h2):
    acc = h1.history["accuracy"] + h2.history["accuracy"]
    val_acc = h1.history["val_accuracy"] + h2.history["val_accuracy"]
    loss = h1.history["loss"] + h2.history["loss"]
    val_loss = h1.history["val_loss"] + h2.history["val_loss"]
    
    epochs_range = range(len(acc))
    
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.plot(epochs_range, acc, label="Train Acc")
    plt.plot(epochs_range, val_acc, label="Val Acc")
    plt.legend()
    plt.title("Accuracy")

    plt.subplot(1,2,2)
    plt.plot(epochs_range, loss, label="Train Loss")
    plt.plot(epochs_range, val_loss, label="Val Loss")
    plt.legend()
    plt.title("Loss")

    plt.show()

plot_history(history1, history2)
```

## 8ï¸âƒ£ Evaluation â€“ Confusion Matrix & Classification Report ğŸ”

We evaluate model performance using:

- **Confusion Matrix**
- **Classification Report (Precision, Recall, F1)**
- **Visual inspection of errors**

This shows how well the fine-tuned MobileNetV2 generalizes to new images.


```python
# Get true labels
y_true = val_gen.classes

# Predict probabilities
y_prob = model.predict(val_gen)

# Convert to class indices
y_pred = np.argmax(y_prob, axis=1)

print("Done predicting!")
```

```python
print("\nğŸ“„ Classification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names))
```

```python
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names,
            yticklabels=class_names)

plt.title("Confusion Matrix")
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()

# Save plot
os.makedirs("plots", exist_ok=True)
plt.savefig("plots/confusion_matrix.png")
plt.show()
```

```python
# find misclassified samples
misclassified = [i for i in range(len(y_true)) if y_true[i] != y_pred[i]]

print(f"Found {len(misclassified)} misclassified images.")

# show 6 examples
plt.figure(figsize=(12,8))

for i, idx in enumerate(random.sample(misclassified, min(6, len(misclassified)))):
    img_path = val_gen.filepaths[idx]
    img = load_img(img_path, target_size=IMG_SIZE)
    
    plt.subplot(2, 3, i+1)
    plt.imshow(img)
    plt.title(f"True: {class_names[y_true[idx]]}\nPred: {class_names[y_pred[idx]]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## 9ï¸âƒ£ Grad-CAM ğŸŒ¡ï¸ â€” Model Interpretability

We generate Grad-CAM heatmaps to visualize **what regions of the image MobileNetV2 focuses on** when making predictions.


```python
os.makedirs("gradcam", exist_ok=True)

def build_grad_model(model,
                     base_name="mobilenetv2_1.00_160",
                     last_conv_name="block_16_project"):
    """
    Builds a grad_model that outputs:
      1) last conv feature maps (from MobileNetV2)
      2) final predictions (through the SAME head)
    This keeps everything in one graph -> gradients work.
    """
    # 1) grab nested base (MobileNetV2)
    base = model.get_layer(base_name)

    # 2) two outputs from base:
    #    - last conv maps
    #    - base final feature tensor
    last_conv_output = base.get_layer(last_conv_name).output
    base_multi = tf.keras.Model(base.input, [last_conv_output, base.output])

    # 3) reuse existing head layers from your model
    gap = model.get_layer("global_average_pooling2d")
    drop = model.get_layer("dropout")
    dense = model.get_layer("dense")

    # 4) wire a single graph
    inputs = model.input
    conv_maps, base_out = base_multi(inputs)

    x = gap(base_out)
    x = drop(x, training=False)
    preds = dense(x)

    grad_model = tf.keras.Model(inputs, [conv_maps, preds])
    return grad_model


def get_gradcam_heatmap(model, img_path, class_idx=None,
                        base_name="mobilenetv2_1.00_160",
                        last_conv_name="block_16_project"):

    # load & preprocess
    img = tf.keras.utils.load_img(img_path, target_size=IMG_SIZE)
    img_array = tf.keras.utils.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    grad_model = build_grad_model(model, base_name, last_conv_name)

    with tf.GradientTape() as tape:
        conv_maps, preds = grad_model(img_array)
        if class_idx is None:
            class_idx = tf.argmax(preds[0])
        loss = preds[:, class_idx]

    grads = tape.gradient(loss, conv_maps)

    # global-average-pool gradients
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_maps = conv_maps[0]
    pooled_grads = pooled_grads

    # weight channels
    conv_maps = conv_maps * pooled_grads

    # mean over channels -> heatmap
    heatmap = tf.reduce_mean(conv_maps, axis=-1)

    # normalize 0..1
    heatmap = tf.maximum(heatmap, 0) / (tf.reduce_max(heatmap) + 1e-10)
    return heatmap.numpy()


def save_gradcam(img_path, heatmap, out_path="gradcam/example.png", alpha=0.45):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    overlay = cv2.addWeighted(img, 1 - alpha, heatmap, alpha, 0)
    cv2.imwrite(out_path, cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))

    plt.figure(figsize=(6,6))
    plt.imshow(overlay)
    plt.axis("off")
    plt.title("Grad-CAM")
    plt.show()
```

```python
# random val image
img_path = random.choice(val_gen.filepaths)
true_label = class_names[val_gen.classes[val_gen.filepaths.index(img_path)]]

print("Image:", img_path)
print("True label:", true_label)

heatmap = get_gradcam_heatmap(
    model,
    img_path,
    base_name="mobilenetv2_1.00_160",
    last_conv_name="block_16_project"
)

save_gradcam(img_path, heatmap, out_path="gradcam/example.png")
print("Saved to gradcam/example.png")
```

```python
plt.figure(figsize=(12,8))

for i in range(6):
    img_path = random.choice(val_gen.filepaths)

    heatmap = get_gradcam_heatmap(
        model,
        img_path,
        base_name="mobilenetv2_1.00_160",
        last_conv_name="block_16_project"
    )

    # overlay to array (no save)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    hm = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    hm = np.uint8(255 * hm)
    hm = cv2.applyColorMap(hm, cv2.COLORMAP_JET)
    overlay = cv2.addWeighted(img, 0.6, hm, 0.4, 0)

    plt.subplot(2,3,i+1)
    plt.imshow(overlay)
    plt.axis("off")

plt.tight_layout()
plt.savefig("gradcam/gradcam_grid.png")
plt.show()

print("Saved grid: gradcam/gradcam_grid.png")


```
