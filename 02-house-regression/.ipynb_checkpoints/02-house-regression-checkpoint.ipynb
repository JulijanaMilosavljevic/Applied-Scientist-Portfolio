# üìå House Prices Prediction ‚Äî Regression Project

This project uses the Kaggle House Prices dataset to predict house sale prices using regression models.
It demonstrates essential end-to-end ML steps required for Applied Scientist interviews.

We will:

* Explore the dataset

* Perform preprocessing & feature engineering

* Train baseline and advanced models

* Evaluate performance

* Save the best model

## üì• Load Dataset

We have two files:

* ```train.csv```

* ```test.csv```

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import xgboost as xgb_module
import matplotlib.pyplot as plt
from math import sqrt
import pickle


train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

train.shape, test.shape
````

## üëÄ Initial Overview

````python
train.head()
````

````python
train.info()
````

````python
train.describe()
````

## ‚ùó Missing Values Analysis

````python
missing = train.isnull().sum().sort_values(ascending=False)
missing[missing > 0]
````

## üõ† Preprocessing Strategy
We will:

* Separate target (```SalePrice```)

* Concatenate train + test for consistent preprocessing

* Impute missing values (median / most frequent)

* Encode categorical features (OneHotEncoder)

* Scale numerical features

````python
y = train["SalePrice"]
X = train.drop("SalePrice", axis=1)

full = pd.concat([X, test], axis=0, ignore_index=True)
full.shape
````

## üîß Feature Engineering
* Fill missing numerical values using median

* Fill categorical missing as **"None"**

* Convert categorical features to strings

* Optionally create new features (Age, TotalArea)

````python
num_cols = full.select_dtypes(include=['int64', 'float64']).columns
cat_cols = full.select_dtypes(include=['object']).columns

full[num_cols] = full[num_cols].fillna(full[num_cols].median())
full[cat_cols] = full[cat_cols].fillna("None")

full.dtypes.value_counts()
````

## üîÑ Encoding Categorical Features

````python
full = pd.get_dummies(full, drop_first=True)
full.shape
````

## üî™ Split Back to Train & Test

````python
X_processed = full.iloc[:len(train)]
test_processed = full.iloc[len(train):]
````

## üß™ Train / Validation Split

````python
X_train, X_val, y_train, y_val = train_test_split(
    X_processed, y, test_size=0.2, random_state=42
)
````

## üß± Baseline Model ‚Äî Linear Regression

````python
lr = LinearRegression()
lr.fit(X_train, y_train)

preds = lr.predict(X_val)
rmse = sqrt(mean_squared_error(y_val, preds))
rmse
````

## üå≤ Random Forest Model

````python
rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)
rf.fit(X_train, y_train)
rf_rmse = sqrt(mean_squared_error(y_val, rf.predict(X_val)))
rf_rmse
````

## üöÄ XGBoost Model (Expected Best Performer)

````python
xgb = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

xgb.fit(X_train, y_train)
xgb_rmse = sqrt(mean_squared_error(y_val, xgb.predict(X_val)))
xgb_rmse
````

## üìä Model Comparison

````python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest", "XGBoost"],
    "RMSE": [rmse, rf_rmse, xgb_rmse]
})
results
````

## ‚≠ê Feature Importance (XGBoost)

````python
xgb_module.plot_importance(xgb, max_num_features=20)
plt.show()
````

## üíæ Save Best Model

````python
pickle.dump(xgb, open("best_model.pkl", "wb"))
````

## üéØ Final Predictions on Test Set

````python
test_preds = xgb.predict(test_processed)

submission = pd.DataFrame({
    "Id": test["Id"],
    "SalePrice": test_preds
})

submission.to_csv("submission.csv", index=False)
````
